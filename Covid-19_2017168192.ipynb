```python
#1.4.9
covid["Active"] = covid["Confirmed"]-covid["Deaths"]-covid["Recovered"]
covid
```

```python
#1.4.10
covidSA = covid[covid["Country/Region"]=="South Africa"]
covidSA['Month'] = covidSA["ObservationDate"].str[0:2]
covgroup = covidSA.groupby('Month').mean()
```

```python
covgroup['Active'].plot(kind = 'line',style = 'b',alpha = 0.4, title = "Average active cases in South Africa per month")
```
